```python
def extract(filename):
    
    tree = ET.parse(filename)

    root = tree.getroot()
    

    img_name = root.find('filename').text
    img_name

    width = float(root.find('size').find('width').text)
    height = float(root.find('size').find('height').text)
    objs = root.findall('object')

    parser = []
    
    for obj in objs:
        name = obj.find('name').text
        bndbox  = obj.find('bndbox')
        xmin =  float(bndbox.find('xmin').text)
        xmax =  float(bndbox.find('xmax').text)
        ymin =  float(bndbox.find('ymin').text)
        ymax =  float(bndbox.find('ymax').text)

        parser.append([img_name, width, height, name,xmin,xmax,ymin,ymax])


    return parser
    
    
    
```

```python
# for detecting unexpected character in files from folder

import os
import string

def has_unexpected_characters(file_path):
    try:
        with open(file_path, 'r', encoding='utf-8') as file:
            contents = file.read()
            # Check if the file contains non-printable characters
            printable_chars = set(string.printable)
            return any(char not in printable_chars for char in contents)
    except UnicodeDecodeError:
        return True  # File is not UTF-8 encoded

folder_path = '/Users/harshalrajput/Desktop/Tensorflow/archive(2)/VOC2012/Annotations'
files_with_unexpected_chars = [
    file for file in os.listdir(folder_path) 
    if has_unexpected_characters(os.path.join(folder_path, file))
]

print("Files with Unexpected Characters:", files_with_unexpected_chars)
```

```python
# for deleting the DS_Store type of file from folder

import os

def delete_ds_store_file(folder_path):
    ds_store_file_path = os.path.join(folder_path, '.DS_Store')
    if os.path.exists(ds_store_file_path):
        os.remove(ds_store_file_path)
        print(f".DS_Store file deleted at: {ds_store_file_path}")
    else:
        print(".DS_Store file not found in the folder.")

folder_path = '/Users/harshalrajput/Desktop/Tensorflow/archive(2)/VOC2012/Annotations'
delete_ds_store_file(folder_path)
```

```python
# move image and label into the respective folder

def save_data(filename, folder_path, groupby_obj):
    
    #move images
    src = os.path.join('/Users/harshalrajput/Desktop/Tensorflow/archive(2)/VOC2012/JPEGImages/'+filename)  # sourse of file
    des = os.path.join(folder_path,filename)  # destination for moving file
    
    shutil.move(src,des)  # move images to destination folder
    
    # save labels
    text_filename = os.path.join(folder_path, os.path.splitext(filename)[0]+'.txt')  # name for txt file
    #os.path.splitext remove extension from filename i.e jpg
    groupby_obj.get_group(filename).set_index('filename').to_csv(text_filename,sep = " ",index = False,header = False)
    
    
    
```

```python
filename_series = pd.Series(groupby_obj_train.groups.keys())  # series of all names of file
```

```python
filename_series.apply(save_data, args = (train_folder, groupby_obj_train))
```

```python
filename_series_test = pd.Series(groupby_obj_test.groups.keys())
```

```python
filename_series_test.apply(save_data, args = (test_folder, groupby_obj_test))
```
